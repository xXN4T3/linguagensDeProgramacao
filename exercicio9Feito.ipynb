## 4. Missão 1 — Barras: Receita por Canal

A missão prática da aula pede construir um gráfico de barras para responder:
**qual canal vende mais?**  
Os slides também sugerem usar o hover para injetar métricas secundárias sem poluir a tela. fileciteturn7file0

### Tarefa
1. Agregue a `receita` por `canal_venda`
2. Inclua também uma métrica secundária, como `quantidade`
3. Construa um gráfico de barras com Plotly Express
4. Ordene do maior para o menor
5. Faça o hover mostrar mais do que apenas a receita

### Perguntas
- Qual canal lidera a receita?
- O canal líder também lidera em quantidade?
- O hover ajudou a enriquecer a leitura sem poluir a tela?


```python
# 1. Agregue a receita e quantidade por canal_venda
df_canal = df.groupby('canal_venda').agg(receita=('receita', 'sum'), quantidade=('quantidade', 'sum')).reset_index()

# 4. Ordene do maior para o menor pela receita
df_canal_sorted = df_canal.sort_values(by='receita', ascending=False)

# 3. Construa um gráfico de barras com Plotly Express
# 5. Faça o hover mostrar mais do que apenas a receita (incluindo quantidade)
fig = px.bar(df_canal_sorted,
             x='canal_venda',
             y='receita',
             title='Receita Total por Canal de Venda',
             labels={'canal_venda': 'Canal de Venda', 'receita': 'Receita Total'},
             hover_data={'quantidade': True, 'receita': ':.2f'})

fig.update_layout(xaxis_title='Canal de Venda', yaxis_title='Receita Total (R$)')
fig.show()
```

```python
# 1. Converta data_venda em data, se necessário (já feito no passo 2)
# df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

# 2. Agregue a receita por mês
df_mensal = df.set_index('data_venda').resample('M')['receita'].sum().reset_index()
df_mensal['mes_ano'] = df_mensal['data_venda'].dt.strftime('%Y-%m') # Para exibir no hover

# 3. Construa um gráfico de linha interativo
fig = px.line(df_mensal,
              x='data_venda',
              y='receita',
              title='Receita Mensal Total ao Longo do Tempo',
              labels={'data_venda': 'Data da Venda', 'receita': 'Receita Total (R$)'},
              hover_name='mes_ano',
              hover_data={'data_venda': '|%Y-%m-%d', 'receita': ':.2f'})

fig.update_xaxes(dtick='M1', tickformat='%b\n%Y') # Formata o eixo X para mostrar mês e ano
fig.update_layout(xaxis_title='Mês', yaxis_title='Receita Total (R$)')
fig.show()
```

Um slide central da aula apresenta três componentes principais:

Escolha um dos seus gráficos interativos e descreva, em markdown:

2. Como o zoom melhora a investigação
3. Como o clique na legenda pode ajudar a isolar séries ou categorias


### Respostas à Tríade da Interatividade (Exemplo: Gráfico de Dispersão — Lucro vs. Receita)

1.  **O que o hover acrescenta?**
    O `hover` transforma cada ponto do gráfico, que é uma representação abstrata de uma transação, em um **caso investigável**. Ao passar o mouse sobre um ponto, conseguimos ver detalhes específicos daquela transação, como o `produto`, `canal_venda`, `uf` e a `margem_lucro`. Isso é crucial para identificar e entender anomalias. Por exemplo, se um ponto mostra alta receita com baixo lucro, o hover permite saber qual produto, em qual UF e canal, contribuiu para essa situação, fornecendo dados concretos para uma investigação posterior.

2.  **Como o zoom melhora a investigação?**
    O `zoom` permite focar em regiões específicas do gráfico, especialmente útil em gráficos com muitos pontos ou onde a maior parte dos dados está concentrada. No gráfico de dispersão, ele nos ajuda a:
    *   **Distinguir pontos próximos:** Em áreas de alta densidade de dados, o zoom revela pontos individuais que poderiam parecer um único aglomerado.
    *   **Isolar anomalias:** Após identificar uma área com potenciais pontos fora da curva, o zoom permite examiná-los com mais clareza, facilitando o uso do hover para obter detalhes.
    *   **Analisar padrões em subgrupos:** Podemos ampliar uma região para ver se há correlações ou comportamentos específicos dentro de um certo intervalo de receita ou lucro.

3.  **Como o clique na legenda pode ajudar a isolar séries ou categorias?**
    O clique na legenda (neste caso, a `categoria` dos produtos) é um **filtro visual** poderoso. Ele permite:
    *   **Focar em categorias específicas:** Clicar em uma categoria na legenda (por exemplo, "Eletrônicos") isola apenas os pontos pertencentes a essa categoria, ocultando os demais. Isso ajuda a entender o comportamento de lucro e receita *dentro* daquela categoria.
    *   **Comparar categorias:** Clicar e desclicar categorias permite comparar rapidamente o perfil de lucro/receita entre elas, identificando se uma categoria específica tem margens consistentemente mais altas ou mais baixas, ou se apresenta mais anomalias. É uma forma rápida de "testar hipóteses" visuais sobre o desempenho de cada segmento.

### Respostas sobre Plotly Express — Máximo impacto, mínimo código

1.  **O que significa dizer que Plotly Express é “declarativo”?**
    Dizer que Plotly Express é "declarativo" significa que o analista **descreve o *que* ele quer** ver no gráfico, em vez de detalhar o *como* construir cada elemento gráfico. Você especifica as variáveis (eixos X, Y, cor, tamanho, etc.) e o tipo de gráfico (linha, barra, dispersão), e o Plotly Express se encarrega de toda a complexidade de gerar o código subjacente do Plotly (JSON, JavaScript, HTML). É como dizer: "eu quero um gráfico de barras com receita no eixo Y e canal de venda no eixo X, e quero que a cor seja a categoria", sem precisar se preocupar com os detalhes de como desenhar as barras ou configurar as legendas.

2.  **Em que ele facilita a vida do analista?**
    O Plotly Express facilita a vida do analista de várias maneiras:
    *   **Menos código, mais resultado:** Com poucas linhas de código, é possível criar gráficos interativos e visualmente atraentes, economizando tempo e esforço.
    *   **Abstração da complexidade:** Ele esconde a complexidade da API principal do Plotly (Graph Objects), permitindo que o analista se concentre mais na análise dos dados do que na sintaxe da visualização.
    *   **Interatividade nativa:** `Hover`, `zoom`, `pan` e `filtros de legenda` são incluídos por padrão, o que é essencial para exploração de dados sem configurações adicionais complexas.
    *   **Integração com Pandas:** É projetado para funcionar perfeitamente com DataFrames do Pandas, o que é a base da manipulação de dados em Python.

3.  **Como isso se conecta com a ideia de futuro dashboard em Streamlit?**
    A conexão do Plotly Express com a ideia de um futuro dashboard em Streamlit é muito forte e vantajosa:
    *   **Componentes reutilizáveis:** Os gráficos interativos criados com Plotly Express podem ser facilmente incorporados em aplicações web construídas com Streamlit. O código que gera um `fig = px.bar(...)` no notebook é quase o mesmo que será usado em um script Streamlit.
    *   **Prototipagem rápida:** Permite criar protótipos de dashboards rapidamente. O analista pode testar visualizações interativas no notebook e, se funcionarem bem, migrá-las para um dashboard Streamlit com pouquíssima modificação.
    *   **Interatividade na web:** Streamlit é uma framework para criar aplicações web interativas com Python. A interatividade que o Plotly Express oferece (zoom, pan, hover) é mantida quando o gráfico é renderizado em um aplicativo Streamlit, proporcionando uma experiência rica para o usuário final do dashboard.
