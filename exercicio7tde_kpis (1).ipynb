## 4. KPIs globais da operação

- **Receita Total**
- **Lucro Total**
- **Margem de Lucro (%)**
- **Ticket Médio**

```python
# Calcule aqui os 4 KPIs globais

receita_total = df['receita'].sum()
lucro_total = df['lucro'].sum()
margem_lucro_percentual = (lucro_total / receita_total) * 100
ticket_medio = receita_total / df['pedido_id'].nunique()

print(f"Receita Total: R$ {receita_total:,.2f}")
print(f"Lucro Total: R$ {lucro_total:,.2f}")
print(f"Margem de Lucro: {margem_lucro_percentual:.2f}%")
print(f"Ticket Médio: R$ {ticket_medio:,.2f}")

```

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?

São Paulo, Rio de Janeiro, Pernambuco e Minas Gerais respectivamente

2. Existe concentração regional?

O maior número se concentra na região sudeste exceto por pernambuco

3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?

devido a alta demanda de São paulo e Rio de Janeiro seria mais coerente reforçar a logistica nessas regiões


```python

analise_uf = df.groupby('uf').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique') # Contagem de pedidos únicos
).reset_index()

# Calcular Margem Calculada (not explicitly asked for, but useful)
analise_uf['margem_calculada'] = (analise_uf['lucro_total'] / analise_uf['receita_total']) * 100

# Calcular Ticket Médio
analise_uf['ticket_medio'] = analise_uf['receita_total'] / analise_uf['quantidade_pedidos']

# Calcular Participação percentual da receita
total_receita_global = df['receita'].sum()
analise_uf['participacao_receita_percentual'] = (analise_uf['receita_total'] / total_receita_global) * 100

# Formatar para melhor visualização
analise_uf['receita_total'] = analise_uf['receita_total'].map('R$ {:,.2f}'.format)
analise_uf['lucro_total'] = analise_uf['lucro_total'].map('R$ {:,.2f}'.format)
analise_uf['margem_calculada'] = analise_uf['margem_calculada'].map('{:,.2f}%'.format) # Optional
analise_uf['ticket_medio'] = analise_uf['ticket_medio'].map('R$ {:,.2f}'.format)
analise_uf['participacao_receita_percentual'] = analise_uf['participacao_receita_percentual'].map('{:,.2f}%'.format)

display(analise_uf.sort_values(by='receita_total', ascending=False))
```
